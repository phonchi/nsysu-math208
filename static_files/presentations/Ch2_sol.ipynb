### (4-1) Given the following code fragment what is its Big-O running time?

```
def main():
    i = n
    count = 0
    while i > 0:
        count = count + 1
        i = i // 2
    return count  
```
(A) $O(n)$

(B) $O(n^2)$

(C) $O(\log n)$

(D) $O(n^3)$

(C)
