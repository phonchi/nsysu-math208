### (5) What is a Sparse Matrix and why we do not use compact array to store it?

Ans: A sparse matrix is a matrix predominantly filled with zero values, distinguishing it from a dense matrix, where most of the elements are non-zero. The rationale for not using compact arrays, which are traditional linear arrays to store sparse matrices, stems from considerations of memory efficiency and computational performance. Storing a sparse matrix in a compact array format would require allocating memory for each element, including the zeros, which is highly inefficient given that the vast majority of the elements do not carry useful information.

### (6) What is the Big O complexity to remove an element from an array?

Ans: The Big O complexity to remove an element from an array depends on the position of the element being removed:

Worst Case (removing the first element or an element at a specific position): 
$O(n)$, where $n$ is the number of elements in the array. This is because every element after the removed element needs to be shifted one position to fill the gap, which requires iterating through the rest of the array.

### (7) We have stored the two-dimensional array students in the memory. The array is 10 × 50 (10 rows and 50 columns). Show the address of the element `students[5][3]` assuming that the element `student[0][0]` is stored in the memory location with address 10 and each element occupies only one memory location. The computer uses column-major storage.

Ans: We  can  use  the  following  formula  to  find  the  location  of  an  element,  assuming  each  element occupies one memory location: 

$y = x + Rows \times j + i$

where `x` defines the start address, Cols defines the number of columns in the array, `i` defines the row number of the element, `j` defines the column number of the element, and `y` is 
address we are looking for. In our example, `x` is 10, `Rows` is 10, `i` is 5 and `j` is 3. We are looking for the value of `y`:

$y = 10 + 10 \times 3 + 5 = 45$

### (8) What is COOrdinate (COO) format?

Ans: The Coordinate (COO) format is a method used to store sparse matrices efficiently. Unlike traditional dense matrix storage that requires space for every element, COO stores only the non-zero elements. It does this by maintaining three arrays: one for the row indices, one for the column indices, and one for the values of the non-zero elements. Each entry in these arrays corresponds to a non-zero element in the matrix, with the row and column arrays indicating the position of the element, and the value array holding the element's actual value. This format is particularly useful for matrices with a large number of zeros because it significantly reduces memory usage.
