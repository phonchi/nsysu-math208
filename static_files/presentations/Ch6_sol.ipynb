### (4-2) How many recursive calls would the call `Fibonacci(5)` generate in the following program?

Fibonacci(3) (second call) calls Fibonacci(2) and Fibonacci(1)

Fibonacci(2) (from the call of Fibonacci(4)) returns 1 directly (3 instances of this)

Fibonacci(1) returns 1 directly (4 instances of this)

To count the total number of calls, including the initial call:

1 for Fibonacci(5) itself

2 for Fibonacci(4) and Fibonacci(3) from the first step

2 more for the calls within Fibonacci(4), which are Fibonacci(3) and Fibonacci(2)

2 for the first instance of Fibonacci(3) calling Fibonacci(2) and Fibonacci(1)

2 more for the second instance of Fibonacci(3) calling Fibonacci(2) and Fibonacci(1)

### (6) What is the main difference between using stack and recursion to write a program?

Ans: 

The main difference between using a stack and recursion in programming lies in their approach to function calls and execution flow:

- Recursion involves a function calling itself to solve smaller instances of the same problem, automatically using the call stack to keep track of its execution state. This can lead to simpler, more readable code for naturally recursive problems but risks stack overflow errors for deep recursions or large inputs due to high memory usage.

- Stack usage in programming typically refers to explicitly managing a stack data structure to track execution state or backtracking without function calls. This method offers more control over memory and can avoid stack overflow issues, but it might result in more complex and less readable code.

### (7) What is the main difference between recursion and iterative programming approaches?

Ans:

The main difference between using recursion and an iterative approach to write a program lies in how they repeat a set of actions and manage their execution flow:

- Recursion involves a function calling itself one or more times in order to solve a smaller part of the problem until it reaches a base case. This approach can lead to elegant, simple solutions for complex problems, particularly those that can naturally be divided into similar, smaller problems. Recursion implicitly uses the call stack to keep track of each function call and its local variables, which can lead to high memory usage and potentially a stack overflow error for deep recursive calls or large input sizes.

- Iterative Approach uses loops to repeat a set of operations until a certain condition is met. Iterative solutions often use less memory than recursive ones because they don't add multiple layers to the call stack. Iterative methods can be more straightforward for problems that naturally fit a step-by-step progression, and they typically avoid the risk of stack overflow.
