```

### (10) What is the main difference between AVL tree and Binary search tree?

The main difference between an AVL tree and a Binary Search Tree (BST) lies in how they manage the balance of their nodes to ensure efficient operations.

1. **Binary Search Tree (BST):**
   - A BST is a data structure where each node contains a key greater than all the keys in the node's left subtree and less than those in the node's right subtree.
   - BST does not necessarily maintain balance. The depth of the tree (and hence the time complexity of search, insert, and delete operations) can vary widely depending on the order of insertion. For instance, inserting elements in sorted order into a BST can produce a tree that is essentially a linked list, with a time complexity of $O(n)$ for basic operations.

2. **AVL Tree:**
   - An AVL tree is a type of BST that maintains a stricter balance.
   - To ensure balance, the AVL tree maintains a balance factor for each node, defined as the height of the left subtree minus the height of the right subtree. This balance factor must always be -1, 0, or +1.
   - After every insertion or deletion, the AVL tree performs rotations if necessary to restore the balance factor requirement. These rotations help to ensure that the tree remains balanced, leading to time complexities of $O(log n)$ for search, insert, and delete operations.

In summary, while both AVL trees and BSTs follow the basic principles of binary search trees, AVL trees include mechanisms to ensure the tree remains balanced, thereby guaranteeing better performance for operations in the worst-case scenarios.

### (11) Which of the inorder/preorder or inorder/postorder or preorder/postorder sequences of a binary tree can determine a unique binary tree? 

To determine a unique binary tree, specific pairs of tree traversal sequences are required because no single traversal sequence (inorder, preorder, or postorder) alone can uniquely identify a binary tree. 

1. **Inorder and Preorder**:
   - **Uniqueness**: Yes, the pair of an inorder and preorder traversal can uniquely determine a binary tree. 
   - **Reason**: The inorder traversal provides information about the relative positions of elements (left to right), and the preorder traversal indicates the root-first ordering. Combining these two, you can uniquely identify the structure of the binary tree because the root element from the preorder sequence can split the inorder sequence into left and right subtrees, which recursively can be used to reconstruct the entire tree.

2. **Inorder and Postorder**:
   - **Uniqueness**: Yes, the pair of an inorder and postorder traversal can also uniquely determine a binary tree.
   - **Reason**: Similar to the inorder and preorder combination, here the postorder gives the root at the end of the sequence. This root element divides the inorder sequence into left and right subtrees. The recursive application of this principle to the subtrees allows the entire tree to be reconstructed uniquely.

3. **Preorder and Postorder**:
   - **Uniqueness**: No, the combination of preorder and postorder traversal sequences does not generally uniquely determine a binary tree.
   - **Reason**: Without the inorder sequence, the relative ordering of nodes between left and right subtrees cannot be definitively established. These two traversals can lead to ambiguities, where multiple different tree structures could have the same preorder and postorder sequences.

```
