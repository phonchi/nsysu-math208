### (5) What are the pros and cons of unorder and order linked list?

Ans: 

Unordered linked lists are simpler and allow for efficient insertions and deletions without worrying about element order, making them flexible for applications where order isn't critical. However, they suffer from inefficient searches as you might need to traverse the entire list to find an element, and they lack direct access to elements by index.

Ordered linked lists, on the other hand, offer improved search efficiency and maintain elements in a sorted order, which is useful for ordered data processing. This order comes at the cost of more complex insertions and deletions, as you must find the correct position to maintain the order, introducing additional overhead and reducing flexibility in how data is managed.

### (6) What is the pros and cons of normal linked list and doubly linked list?

Ans:
A normal (singly) linked list is simpler and uses less memory per node since each node only stores a reference to the next node. This makes it efficient for memory usage and straightforward operations like insertion and deletion at the beginning of the list. However, it lacks the ability to traverse backwards, making operations that require reverse traversal or deletion from the end less efficient, as you have to iterate from the head to get to the previous node.

Doubly linked lists, conversely, allow for traversal in both directions thanks to an extra pointer in each node pointing to the previous node. This added functionality facilitates operations like reverse traversal and efficient deletion from both ends but comes at the cost of increased memory usage per node and slightly more complex insertion and deletion operations, as two pointers need to be updated instead of one.
