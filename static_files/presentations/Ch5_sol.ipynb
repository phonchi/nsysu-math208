### (6) Assume we have a stack S implementing the operations defined in the stack Abstract Data Type (ADT). What will be the contents of the stack after executing the following operations? Please identify the topmost element.

```
S = stack()
S.push(10)
S.push(5)
if (not S.is_empty()) S.pop()
S.push(12)
S.push(7)
S.pop()
```

After the first step: [10]

After the second step: [5,10]

After the third step: [10]

After the fourth step: [12,10]

After the fifth step: [7,12,10]

After the sixth step: [12,10]

The content will be [12,10] where 12 is at the top

### (7) Assume we have a stack S and a queue Q, each implementing the operations defined in their respective Abstract Data Type (ADT) standards.
```
Q = queue()
S = stack()
Q.enqueue(15)
Q.enqueue(10)
x = Q.dequeue()
if (not Q.is_empty()) Q.dequeue()
S.push(x)
S.push(2)
S.pop()
S.push(5)
```

After the second step: Q=[15,10]

After the third step: Q=[20], x=15

After the fourth step: Q=[]

After the fifth step: S=[15]

After the sixth step: S=[15,2]

After the seventh step: S=[15]

After the eighth step: S=[15,5]

The content of stack is [15,5] where 5 is at the top and the queue is empty.
